Merging DataFrames

```
from google.colab import files
import numpy as np
import pandas as pd
import glob

uploaded = files.upload()

##files.upload returns a dictionary of the files which were uploaded. The 
##dictionary is keyed by the file name, the value is the data which was 
##uploaded.
i = 1
df_alldata = pd.DataFrame()
all_data = pd.DataFrame()
df_alldata['mdate'] =[]
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(name=fn, length=len(uploaded[fn])))
  df_alldata = pd.read_excel(fn)
  df_alldata['mdate'] = i
  i = i + 1
  all_data = all_data.append(df_alldata,ignore_index=True)
```

Missing Data and its imputation

```
#missing data
total = all_data.isnull().sum().sort_values(ascending=False)
percent = (all_data.isnull().sum()/all_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
print(missing_data)
```
